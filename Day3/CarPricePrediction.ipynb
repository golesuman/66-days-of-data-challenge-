```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',header=None)
df.head()
```

```python
df.columns=['ID','normalized-losses','make','fuel-type','aspiration','num-of-doors','body_style','drive_wheels','engine-location','wheel_base','length','width','height','curb-weight','engine type','num-of-cylinders',
           'engine_size','fuel_system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```python
df.head()
```

```python
# checking the data shape info
df.info()
```

```python
df.shape
```

```python
# checking the no of null datasets in the data
```

```python
df.isnull().sum()
```

```python
df.replace({'?':np.nan},inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['normalized-losses','num-of-doors','bore','stroke','horsepower','peak-rpm','price'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['normalized-losses']=df['normalized-losses'].astype(int)
df['price']=df['price'].astype(float)
df['peak-rpm']=df['peak-rpm'].astype(int)
df['horsepower']=df['horsepower'].astype(int)
df['compression-ratio']=df['compression-ratio'].astype(float)
df['bore']=df['bore'].astype(float)
df['stroke']=df['stroke'].astype(float)
df['normalized-losses'].dtypes
```

```python
df.head()
```

```python
df.info()
```

```python
category=[feature for feature in df.columns if df[feature].dtypes=='O']
category
```

## Let's plot the box plot to see the distribution and the percentiles of the categorical data

```python
plt.figure(figsize=(12,8))

for feature in category:
    sns.boxplot(x=feature,y='price',data=df.sort_values('price'))
    plt.show()
```

### Lets plot the countplot for the categorical variables


```python
plt.figure(figsize=(15,8))
for feature in category:
    sns.countplot(df[feature])
    plt.show()
```

```python
conti=[feature for feature in df.columns if len(df[feature].unique())>25]
conti
```

## Let's see if the given continuous data are normally distributed or not

```python
plt.figure(figsize=(12,8))
for feature in conti:
    sns.distplot(df[feature])
    plt.show()
```

All the data seems to be normally distributed and  now we can do feature engineering 

```python
# Let's change the categorical variable to the numerical value
category
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
for feature in category:
    df[feature]=lbl.fit_transform(df[feature])
```

```python
df.head()
```

```python
# since the id is not important we can drop that feature 
df.drop('ID',axis=1,inplace=True,)
```

```python
df.head()
```

### Let's see the correlation between the features and the price

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(),annot=True)
```

```python
## let's drop some features having low correlation with the price

x=df.drop(['make','price','city-mpg','peak-rpm'],axis=1)
```

```python
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import AdaBoostRegressor
model=AdaBoostRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
score=model.score(x_test,y_test)
score
```

```python
y_pred=model.predict(x_test)
y_pred
```
