```python
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
df['ocean_proximity']=encode.fit_transform(df['ocean_proximity'])
```

Since all the continuous data looks distributed normally we can now do some feature engineering


```python
x=df.drop('Price',axis=1)
y=df.Price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3,test_size=0.2)
from sklearn.linear_model import LassoLars,Ridge
clf=LassoLars(alpha=1.0)
clf2=Ridge(alpha=1.0)
clf2.fit(x_train,y_train)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
from sklearn.metrics import r2_score
score=r2_score(y_true=y_test,y_pred=y_pred)
score
```

```python
y2_pred=clf2.predict(x_test)
score2=r2_score(y_true=y_test,y_pred=y2_pred)
score2
```
