```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('pizza_bluescreen.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
lower_blue=np.array([0,0,220])
upper_blue=np.array([50,70,255])
#creating mask
mask=cv2.inRange(img,lower_blue,upper_blue)
plt.imshow(mask,cmap='gray')
```

```python
masked_image=np.copy(img)
masked_image[mask!=0]=[0,0,0]
plt.imshow(masked_image)
```

```python
backround=cv2.imread('/content/space_background.jpg')
backround_image=cv2.cvtColor(backround,cv2.COLOR_BGR2RGB)
crop_backround=backround_image[0:514,:816]
plt.imshow(crop_backround)
```

```python
crop_backround[mask==0]=[0,0,0]
plt.imshow(crop_backround)
```

```python
complete_image=crop_backround+masked_image
plt.imshow(complete_image)
```
